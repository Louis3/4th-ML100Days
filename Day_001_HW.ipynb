```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Mean_sequare_error(y,yi):
    """
    計算MSE 均方誤差
    yi:真實值
    y :預測值
    Return : 
    mse : Mean sequare error
    """
    mse = sum((y - yi)**2) / len(y)
    return mse

w = 3
b = 0.5
x_lin = np.linspace(0, 100, 101)
y = (x_lin + np.random.randn(101) * 5) * w + b

plt.plot(x_lin, y, 'r.', label = 'data points')
plt.title("Assume we have data points")
plt.legend(loc = 2)
plt.show()

y_hat = x_lin * w + b
plt.plot(x_lin, y, 'r.', label = 'data')
plt.plot(x_lin, y_hat, 'b-', label = 'prediction')
plt.title("Assume we have data points (And the prediction)")
plt.legend(loc = 2)
plt.show()

MSE = Mean_sequare_error(y, y_hat)
print("The Mean squared error is %.3f" % (MSE))
```
